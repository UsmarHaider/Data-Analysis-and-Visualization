```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("bird-window-collision-death.csv")
df
```

```python
fig = px.sunburst(df, path=['Bldg #', 'SpeciesCode'], values='Deaths')
fig.update_layout(legend_itemclick=False, legend_itemdoubleclick=False)
fig.show()
fig.write_html('20l-1095.html')
```

```python
fig = px.histogram(df, x='Bldg #', y='Deaths', color='SpeciesCode')
fig.update_layout(legend_itemclick=False, legend_itemdoubleclick=False)
fig.show()
fig.write_html('20l-1095.html')
```
