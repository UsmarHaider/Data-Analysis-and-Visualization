```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
orignal_data=pd.read_excel('Original Iris.xlsx')
training_data=pd.read_excel('my_iris.xlsx')
```

```python
orignal_data
```

```python
orignal_data.drop(orignal_data.columns[0],axis=1, inplace=True)
```

```python
orignal_data.info()
```

```python
orignal_data.corr()
```

```python
TData_1 = LabelEncoder()
orignal_data['target'] = TData_1.fit_transform(orignal_data['target'])
orignal_data.head()
```

```python
test1 = orignal_data.drop(orignal_data.columns[-1], axis=1 )
test2= orignal_data[orignal_data.columns[-1]]
x_training_1, x_test_1, y_training_1, y_test_1 = train_test_split( test1 , test2 , test_size=0.2 , random_state=0)
mod_1 = LogisticRegression(max_iter=2000)
```

```python
mod_1.fit(x_training_1, y_training_1)
```

```python
result=mod_1.score(x_test_1, y_test_1) * 100
print("Correction:)", result)
```
