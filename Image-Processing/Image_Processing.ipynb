```python
def grayscale(img):
    R, G, B = img[:, :, 0], img[:, :, 1], img[:, :, 2]
    grayscale = ((0.3 * R) + (0.59 * G) + (0.11 * B))
    return grayscale
```

```python
# Read in the cat images
for i in range(0, 6):
    img = mpimg.imread('cat' + str(i) + '.jpg')
    imgGray = grayscale(img)
    cv2.imwrite('greycat.' + str(i) + '.jpg', imgGray)
    plt.imshow(imgGray, cmap='gray')
    plt.show()
```

```python
def avgfilter(imgGray):

    m, n = imgGray.shape
    #Develop Averaging filter(3, 3) mask
    mask = np.ones([3, 3], dtype = int)
    mask = mask / 9
  #Convolve the 3X3 mask over the image
    img_new = np.zeros([m, n])


    for i in range(1, m-1):

        for j in range(1, n-1):

            temp = imgGray[i-1, j-1]*mask[0, 0]+imgGray[i-1, j]*mask[0, 1]+imgGray[i-1, j + 1]*mask[0, 2]+imgGray[i, j-1]*mask[1, 0]+ imgGray[i, j]*mask[1, 1]+imgGray[i, j + 1]*mask[1, 2]+imgGray[i + 1, j-1]*mask[2, 0]+imgGray[i + 1, j]*mask[2, 1]+imgGray[i + 1, j + 1]*mask[2, 2]
            img_new[i, j]= temp
    img_new = img_new.astype(np.uint8)
    #cv2.imwrite('blurred.' + str(i) + '.jpg', img_new)
    plt.imshow(img_new, cmap='gray')
    plt.show()
```

```python
# Average Filter in the cat images
for i in range(0, 6):
    img = mpimg.imread('greycat.' + str(i) + '.jpg')
    imgGray = avgfilter(img)

```

```python
def medfilter(imgGray):

#Median Filter
    img_noisy1=imgGray
    m, n = img_noisy1.shape

 # Traverse the image. For every 3X3 area,
# find the median of the pixels and
# replace the center pixel by the median
    img_new1 = np.zeros([m, n])

    for i in range(1, m-1):
        for j in range(1, n-1):
            temp = [img_noisy1[i-1, j-1],
               img_noisy1[i-1, j],
               img_noisy1[i-1, j + 1],
               img_noisy1[i, j-1],
               img_noisy1[i, j],
               img_noisy1[i, j + 1],
               img_noisy1[i + 1, j-1],
               img_noisy1[i + 1, j],
               img_noisy1[i + 1, j + 1]]
            temp = sorted(temp)
            img_new1[i, j]= temp[4]

    img_new1 = img_new1.astype(np.uint8)
    plt.imshow(img_new1, cmap='gray')
    plt.show()
```

```python
#median filter
for i in range(0, 6):
    img = mpimg.imread('greycat.' + str(i) + '.jpg')
    img = medfilter(img)
```
