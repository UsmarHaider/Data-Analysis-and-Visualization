```python
def generate_ngrams(text, WordsToCombine=2):
     words = text.split()
     output = []
     for i in range(len(words)- WordsToCombine+1):
         output.append(tuple(words[i:i+WordsToCombine]))
     return output
```

```python
def lower_case(words_list):
    return [words.lower() for words in words_list]
```

```python
def remove_punc(words_list):
    words_list=[''.join(c for c in s if c not in string.punctuation) for s in words_list]
    words_list= filter(lambda word: word != ' ', words_list)
    words_list=list(words_list)
    return words_list
```

```python
def unique_words(sequence):
    words = set()
    for i in range(0,10):
        for j in sequence[i]:
            words.add(j)
    return words
```

```python
number=input("Enter Number of N gram you want to use\n")
number=int(number)
```

```python
sentence=[]
words_n_gram=[]
for i in range(0,10):
    x=[df1[i]]
    x=lower_case(x)
    x=remove_punc(x)
    x=x[0].strip()
    sentence=sentence+[x]
    words_n_gram.append(generate_ngrams(x,number))
```

```python
un_words=unique_words(words_n_gram)
print("Unique Words are : ",len(un_words))
print(un_words)
```

```python
def ngram_frequncy(un_words,words_list):
    vector={}
    for x in un_words:
        if words_list.count(x)>0:
            vector[x]=1
        else:
            vector[x]=0
    return vector
```

```python
n_gram_lis=[]
for i in range(0,10):
    n_gram_lis.append(ngram_frequncy(un_words,words_n_gram[i]))
```

```python
df_n_gram=pd.DataFrame(n_gram_lis,index=pd.DataFrame(sentence))
```
